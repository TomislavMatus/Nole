```python
import PyPDF2 as pdf
import re
from top2vec import Top2Vec
```

```python
file = open('/Users/Tomi/code/TomislavMatus/Nole/data/NoleBio.pdf', 'rb')
```

```python
pdf_reader = pdf.PdfFileReader(file)
```

```python
pageCount = pdf_reader.getNumPages()
```

```python
%%time

def extractText(pageCount, pdf_reader):
    NoleBioExtracted = []
    for i in range (pageCount):
        print(pdf_reader.pages[i].extract_text())
        NoleBioExtracted.append(pdf_reader.pages[i].extract_text())
    return NoleBioExtracted
```

```python
%%time
NoleBioExtracted = extractText(200, pdf_reader)
```

```python
final_list = []
for chapter in NoleBioExtracted:
    if "." in chapter:
        sentences = chapter.split(".")
        for x in sentences:
            final_list.append(x)
    
```

```python
cleaned = []
for sentence in final_list:
    sentence = sentence.replace("\n", "")
    cleaned.append(sentence)
```

```python
model = Top2Vec(cleaned, min_count=100, embedding_model="universal-sentence-encoder")
```
